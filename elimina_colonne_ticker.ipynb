```python
import pandas as pd
import os

df = pd.read_hdf('tickers/TRU.h5', 'ticker')
df.info()
```

```python

lista_files = os.listdir('tickers')
i = 1
tot = len(lista_files)

def elimina_colonna(df, colonna):
    if colonna in df.columns:
        df.drop(columns=[colonna], axis=1, inplace=True)
    return df
    

for file_name in lista_files:
    file_path = os.path.join('tickers', file_name)
    # Controlla se è un file prima di procedere
    if os.path.isfile(file_path):
        print(f'{i} di {tot}) Ticker {file_name}')
        ticker = pd.read_hdf(file_path, 'ticker')
        ticker = elimina_colonna(ticker, 'Perc_Max_High_Futuro_20d')
        ticker = elimina_colonna(ticker, 'Perc_Drawdown_20d')
        ticker = elimina_colonna(ticker, 'Perc_Max_High_Futuro_50d')
        ticker = elimina_colonna(ticker, 'Perc_Drawdown_50d')
        ticker = elimina_colonna(ticker, 'Perc_Max_High_Futuro_100d')
        ticker = elimina_colonna(ticker, 'Perc_Drawdown_100d')
        ticker = elimina_colonna(ticker, 'max_gain')
        ticker = elimina_colonna(ticker, 'max_drawdown')
        ticker = elimina_colonna(ticker, 'MaxMinRel')
        ticker = elimina_colonna(ticker, 'pct_change')
        ticker = elimina_colonna(ticker, 'pct_change_5d')
        ticker = elimina_colonna(ticker, 'pct_change_10d')
        ticker = elimina_colonna(ticker, 'pct_change_20d')
        ticker = elimina_colonna(ticker, 'Min_Close_5d')
        ticker = elimina_colonna(ticker, 'pct_change_5d_uscita')

        i += 1
        ticker.to_hdf(file_path, key='ticker', mode='w')

```
