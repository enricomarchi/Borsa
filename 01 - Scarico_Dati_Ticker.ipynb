```python
import os
import yfinance as yf
import pandas as pd

# Creare la cartella Tickers se non esiste
if not os.path.exists('C:/Users/932197/Documents/dati_locali/Tickers'):
    os.makedirs('C:/Users/932197/Documents/dati_locali/Tickers')

# List of NASDAQ and NYSE tickers. You need to replace this with the actual list of tickers.
df_ticker = pd.read_excel(io="C:/Users/932197/OneDrive/Python/DeGiro con formule.xlsx", header=0)
df_ticker = df_ticker[['Ticker', 'USIN', 'Categoria', 'Borsa']]

tot_tickers = len(df_ticker)

# Load the last ticker processed if the file exists.
last_ticker_processed = None
if os.path.exists('C:/Users/932197/Documents/dati_locali/last_ticker.txt'):
    with open('C:/Users/932197/Documents/dati_locali/last_ticker.txt', 'r') as f:
        last_ticker_processed = f.read().strip()

start_from_next_ticker = last_ticker_processed is None

i = 1
for index, ticker in df_ticker.iterrows():
    if not start_from_next_ticker:
        if ticker["USIN"] == last_ticker_processed:
            start_from_next_ticker = True
        print(f"\r{i} di {tot_tickers}) Caricamento ticker {ticker['Ticker']}                           ", end=' ', flush=True)
        i += 1
        continue
    
    print(f"\r{i} di {tot_tickers}) Downloading {ticker['Ticker']}...               ", end=" ", flush=True)
    try:
        data = yf.download(ticker['USIN'], start='2013-01-01', end='2023-06-30', progress=False)
        if len(data) > 500:
            data.to_parquet(f'C:/Users/932197/Documents/dati_locali/Tickers/{ticker["USIN"]}.parquet')
    except Exception as e:
        print(f"\nError downloading {ticker['Ticker']}: {str(e)}\n")
        continue

    with open('C:/Users/932197/Documents/dati_locali/last_ticker.txt', 'w') as f:
        f.write(ticker["USIN"])
    i += 1

os.remove("C:/Users/932197/Documents/dati_locali/last_ticker.txt")
print("\nDownload complete.")
```
